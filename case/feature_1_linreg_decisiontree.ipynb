```python
import pandas as pd
import numpy as np
import sklearn
```

```python
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('data.csv', sep=";", usecols=["Feature 1", "Feature 2", "Y"])
X, Y = data["Feature 1"].to_numpy(), data["Y"].to_numpy()
X_train, X_test, y_train, y_test  = train_test_split(X, Y, test_size=0.2, random_state=1)
data.head()
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train.reshape(X_train.shape[0], 1), y_train.reshape(X_train.shape[0], 1))
tree.plot_tree(clf)
```

```python
sklearn.metrics.mean_squared_error(y_test, clf.predict(X_test.reshape(y_test.shape[0], 1)))
```

```python
regr = sklearn.linear_model.LinearRegression()
regr.fit(X_train.reshape(X_train.shape[0], 1), y_train.reshape(X_train.shape[0], 1))
regr.coef_
```

```python
sklearn.metrics.mean_squared_error(y_test, regr.predict(X_test.reshape(X_test.shape[0], 1)))
```
